```python
n_components = 2
random_state = 2021

def plot_series(x, y, title):
    fig, ax = plt.subplots()
    fig.set_figheight(8)
    fig.set_figwidth(13)

    sns.lineplot(x=x, y=y, color="blue")

    ax.set_title(title)
    ax.set_xlabel("")
    ax.set_ylabel("Amount")
    plt.plot()
    return ax

def perform_cpa(n_component, random_state):
    pca = PCA(n_components=n_components, random_state=random_state)
    X_pca = pca.fit_transform(X)
    print("Mean: ", np.mean(X_pca), "STD: ", np.std(X_pca))
    
    X_train_inverse = pca.inverse_transform(X_pca)
    X_train_inverse = pca.fit_transform(X_train_inverse)
    
    residual = X_pca - X_train_inverse
        
    print("Residual: ", residual)
        
    return X_pca, X_train_inverse, residual

X_pca, X_train_inverse, residual = perform_cpa(n_components, random_state)

residual_df = pd.DataFrame(data=residual, columns = ['PC1', 'PC2'])
    
sns.lmplot(x="PC1", y="PC2",
      data=residual_df, 
      fit_reg=False, 
      legend=True,
      scatter_kws={"s": 5}) # specify the point size

# plot_series(residual[::, 0], X_pca[::, 0], "Residual")
# plot_series(residual[::, 1], X_pca[::, 1], "Residual")

# plot_series(residual[::, 0], X_train_inverse[::, 0], "Residual")
# plot_series(residual[::, 1], X_train_inverse[::, 1], "Residual")

# plt.scatter(X_pca[8000:10000, 0], X_pca[8000:10000, 1], alpha=0.2)
# plt.scatter(residual[8000:10000, 0], residual[8000:10000, 1], alpha=0.8)
# plt.axis('equal')

# plt.show()

# plt.scatter(X_pca[18000:20000, 0], X_pca[18000:20000, 1], alpha=0.2)
# plt.scatter(residual[18000:20000, 0], residual[18000:20000, 1], alpha=0.8)
# plt.axis('equal')

# plt.show()

# plt.scatter(X_pca[25000:30000, 0], X_pca[25000:30000, 1], alpha=0.2)
# plt.scatter(residual[25000:30000, 0], residual[25000:30000, 1], alpha=0.8)
# plt.axis('equal')

# plt.show()
```

```python
def anomalies(column):
    window_percentage = 5
    k = int(len(column) * (window_percentage/100))
    N = len(column)

    time = np.arange(0,N)

    column = column.to_numpy()

    get_bands = lambda df : (np.mean(df) + 3*np.std(df),np.mean(df) - 3*np.std(df))

    bands = [get_bands(column[range(0 if i - k < 0 else i - k ,i + k if i + k < N else N)]) for i in range(0,N)]
    upper, lower = zip(*bands)

    anomalies = (column > upper) | (column < lower)

    plt.figure(figsize=(20,10))
    plt.plot(time,column,'k',label='Data')
    plt.plot(time,lower,'r-',alpha=0.3)

#     plt.plot(df[anomalies], column[anomalies],'ro',label='Anomalies')
#     plt.fill_between(time, upper, lower,facecolor='red',alpha=0.1)
    plt.legend()

    plt.show()

    
# The following functions take a long time to compute completely. For easier and faster results let them only show (max) 50.000 results
anomalies(df['FIT101'])
# anomalies(df['LIT101'])
# anomalies(df['MV101'])
# anomalies(df['P101'])
# anomalies(df['P102'])
```
